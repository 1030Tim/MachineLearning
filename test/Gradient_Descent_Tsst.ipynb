```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
x = df['YearsExperience']
y = df['Salary']
plt.scatter(x,y,edgecolors='black',color='blue',linewidths=2)
plt.show()
```

```python
# computs_cost
def computs_cost(w,x,b,y)->int:
    y_pred = w*x+b
    cost = ((y-y_pred)**2).mean()
    return cost
```

```python
def computs_gradient(w,x,b,y):
    w_gradient = (x*(w*x+b -y)).mean()
    b_gradient = (w*x+b -y).mean()
    return w_gradient,b_gradient
```

```python
def gradient_descent(w_init,b_init,count,learning_rate):
    w = w_init
    b = b_init

    w_hist = [ ]
    b_hist = [ ]
    c_hist = [ ]

    for i in range(count):
        w_gradient,b_gradinet = computs_gradient(w,x,b,y)
        w = w - w_gradient*learning_rate
        b = b - b_gradinet*learning_rate
        cost = computs_cost(w,x,b,y)
        w_hist.append(w)
        b_hist.append(b)
        c_hist.append(cost)

        if (i%1000 == 0):
            print(f"Intertion {i:5} cost: {cost:.2e} W: {w:.2e} B: {b:.2e} w_gradient {w_gradient} b_gradient {b_gradinet}")

    return w,b,w_hist , b_hist, c_hist
```

```python
learning_rate = 1.0e-6
w_init = 0
b_init = 0
count = 20000

w_final,b_final,w_hist , b_hist , c_hist = gradient_descent(w_init,b_init,count,learning_rate)
```

```python
#interation vs cost
plt.plot(np.arange(0,100),c_hist[:100])
plt.title('interation vs cost')
plt.xlabel('interation')
plt.ylabel('cost')
plt.show()
```

```python
learning_rate = 1.0e-3
w_init = -100
b_init = 0
count = 100000
w_final,b_final,w_hist,b_hist,c_hist = gradient_descent(w_init,b_init,count,learning_rate)
```

```python
ws = np.arange(-100,101)
bs = np.arange(-100,101)
costs = np.zeros((201,201))

i = 0
for w in ws:
    j = 0
    for b in bs:
        cost = computs_cost(w,x,b,y)
        costs[i,j] = cost
        j+=1
    i+=1
costs
```

```python
ax = plt.axes(projection='3d')
ax.xaxis.set_pane_color((1.0,1.0,1.0))
ax.yaxis.set_pane_color((1.0,1.0,1.0))
ax.zaxis.set_pane_color((1.0,1.0,1.0))

b_grid , w_grid = np.meshgrid(bs,ws)
ax.plot_surface(w_grid,b_grid,costs,alpha=0.3)
ax.set_title("w vs b cost")
ax.set_xlabel("w")
ax.set_ylabel("b")
ax.set_zlabel("cost")

w_index,b_index = np.where(costs == np.min(costs))
ax.scatter(ws[w_index],bs[b_index],costs[w_index,b_index],color="red",s = 50)
ax.scatter(w_hist[0],b_hist[0],c_hist[0],color="green",s = 30)
ax.plot(w_hist,b_hist,c_hist)
plt.show()
```
