```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/ShiYu0318/SCIST_2025_WC_AI/main/study_hours_scores.csv"
df = pd.read_csv(url)
x = df["Study_Hours"]
y = df["Scores"]
df
```

```python
plt.scatter(x,y,linewidths=2,edgecolors="black",color = "blue")
plt.show()
```

# 假設模型y = wx+b

```python
def compute_cost(x, y, w, b):
  y_pred = w*x + b
  cost = (y - y_pred)**2
  cost = cost.sum() / len(x)

  return cost
```

```python
# w=-100~100 b=-100~100 的 cost
import numpy as np

ws = np.arange(-100, 101)
bs = np.arange(-100, 101)
costs = np.zeros((201, 201))

i = 0
for w in ws:
  j = 0
  for b in bs:
    cost = compute_cost(x, y, w, b)
    costs[i,j] = cost
    j = j+1
  i = i+1
costs
```

```python
ax = plt.figure().add_subplot(projection='3d')
b_grid, w_grid = np.meshgrid(bs,ws)
ax.plot_surface(b_grid,w_grid,costs,alpha = 0.4)
w_index,y_index = np.where(costs == np.min(costs))
ax.scatter(ws[w_index],bs[y_index],costs[w_index,y_index],color='red')
print(w_index,y_index)
```

```python
print(w_index,y_index)
print(f"當 w = {ws[w_index]},b = {bs[y_index]}會有最小cost:{np.sqrt(costs[w_index,y_index])}")
```

```python
plt.scatter(x,y,edgecolors='black',color='blue')
y_pred = ws[w_index]*x+bs[y_index]
plt.plot(x,y_pred,color='red')
plt.show()
```
