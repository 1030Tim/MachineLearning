```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("scorsrvstime.csv")
x = df['學習時間']
y = df['成績']
```

```python
def cost_function(y,w,x,b):
    y_pred = w*x+b
    costs = (y - y_pred)**2
    cost = costs.sum() /len(x)
    return cost
```

```python
# cost w
costs = [ ]
for w in range(-100,101):
    cost = cost_function(y,w,x,0)
    costs.append(cost)
costs
```

```python
plt.plot(range(0,201),costs)
plt.title('study vs cost')
plt.ylabel('w')
plt.xlabel('cost')
plt.show()
```

```python
# 暴力解
ws = np.linspace(-1,1,100)
bs = np.linspace(-1,1,100)
costs = np.zeros((100,100))

i = 0
for w in ws:
    j = 0
    for b in bs:
        cost = cost_function(y,w,x,b)
        costs[i,j] = cost
        j+=1
    i+=1
costs
```
