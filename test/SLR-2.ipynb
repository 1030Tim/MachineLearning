```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('Study_Hourse_and_Scorse.csv')
x = df['Study_Hours']
y = df['Scores']
```

```python
w = 0
b = 0
def comput_cost(x,y,w,b):
    y_pred = w*x+b
    cost = (y - y_pred)**2
    costs = cost.sum()/len(x)
    return costs
```

```python
costs = [ ]
ws = np.arange(-100,101)
bs = np.arange(-100,101)
costs = np.zeros((201,201))

# 枚舉法
i = 0
for w in ws:
    j = 0
    for b in bs:
        costs[i,j] = comput_cost(x,y,w,b)
        j+=1
    i+=1
costs
```

```python
plt.plot(range(-100,101), costs)

plt.title("cost function")
plt.xlabel("w")
plt.ylabel("cost")
plt.show()
```

```python
ax = plt.figure().add_subplot(projection='3d')
w_gred,b_gred = np.meshgrid(ws,bs)
ax.plot_surface(w_gred,b_gred,costs,alpha=0.4)
w_index,b_index = np.where(costs == np.min(costs))
ax.scatter(ws[w_index],bs[b_index],color='red')
plt.show()
print(f"if min in w {ws[w_index]} b {bs[b_index]}")
```

```python
good_w,good_b = ws[w_index],bs[b_index]
y_pred = good_w*x+good_b
plt.scatter(x,y,edgecolors='black',color='blue')
plt.plot(x,y_pred,color='red')
plt.xlabel("Hours")
plt.ylabel("Scorse")
plt.title("Stady hours vs Scorse")
plt.show()
```
