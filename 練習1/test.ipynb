```python
import pandas as pd
import matplotlib.pylab as plt
import matplotlib as mpl
mpl.rc('font',family='DFYuanLight-B5')
mpl.rcParams['axes.unicode_minus']=False
df = pd.read_csv('練習檔案1.csv')
```

```python
x = df['Hours']
y = df['Score']
plt.scatter(x,y,marker = 'x')
plt.show()
```

```python
#cost function
def comput_cost(x,y,w,b):
    y_pred = w*x+b
    cost = (y-y_pred)**2
    cost = cost.sum()/len(x)
    return cost
```

```python
comput_cost(x,y,10,10)
```

#### w在(-100,100)間的cost

```python
costs = [ ]
for w in range(-100,101):
    cost_function = comput_cost(x,y,w,10)
    costs.append(cost_function)
costs
```

```python
plt.plot(range(-100,101),costs)
plt.xlabel('w')
plt.ylabel('cost')
plt.show()
```

#### w,b range in (-100,100)

```python
import numpy as np
i = 0
ws = np.arange(-100,101)
bs = np.arange(-100,101)
costs = np.zeros((201,201))
for w in ws:
    j = 0
    for b in bs:
        costs[i,j] = comput_cost(x,y,w,b)
        j+=1
    i+=1
costs
```

```python
ax = plt.axes(projection = '3d')
b_grid,w_grid = np.meshgrid(bs,ws)
ax.view_init(40,-100)
ax.plot_surface(w_grid,b_grid,costs,alpha = 0.5)
ax.plot_wireframe(w_grid,b_grid,costs,alpha = 0.5)

ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Cost')
ax.set_title('學習時間與成績')
ax.legend()

w_index,b_index = np.where(costs == np.min(costs) )
ax.scatter(ws[w_index],bs[b_index],costs[w_index,b_index], color='red')

plt.show()
print(f"當 w = {ws[w_index]},b = {bs[b_index]}會有最小cost:{costs[w_index,b_index]}")
```
