```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # 需要這個庫來畫熱力圖

# 假設你的資料已經讀入 df 中
# 如果需要從 CSV 讀取，請取消下面這行的註釋
# df = pd.read_csv("train_dataset.csv")

# 計算相關性矩陣
correlation_matrix = df.corr()

# 設定圖形大小
plt.figure(figsize=(12, 10))

# 繪製熱力圖
sns.heatmap(correlation_matrix, 
            annot=True,          # 顯示數值
            cmap='coolwarm',     # 顏色方案
            center=0,           # 將0設為顏色中心
            square=True,        # 使格子為正方形
            fmt='.2f',          # 數值格式(保留2位小數)
            linewidths=0.5)     # 格子間的線寬

# 設定標題
plt.title('Feature Correlation Heatmap', pad=20, size=16)

# 旋轉x軸標籤以避免重疊
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# 調整佈局
plt.tight_layout()

# 顯示圖形
plt.show()
```
