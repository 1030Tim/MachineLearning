```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train_dataset.csv")
print(df.keys())
df.head()
```

```python
df.isnull().sum() #data if None ?
```

```python
df["MEDE"] = df["PRICE"]
```

```python
sns.displot(data=df, x="MEDE", kde=True) 
plt.show()
```

```python
hote_image = df.corr()
plt.figure(figsize=(12,12))
sns.heatmap(hote_image,
            annot=True,
            cmap="coolwarm",
            center=0,
            square=True,
            fmt=".2f",
            linewidths=0.5)
```
